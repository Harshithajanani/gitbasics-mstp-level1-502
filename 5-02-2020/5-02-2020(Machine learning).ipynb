```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### READING THE DATA
data=pd.read_csv("Advertising.csv")
data.tail()
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
data.tail()
```

```python
x=data["TV"]
x
```

```python
x=data["TV"].values
x
```

```python
x=data["TV"].values
print(x.shape)
x
```

```python
y=data["sales"].values
print(y.shape)
y
```

```python
## Mean calculation
x_mean=np.mean(x)
print("x.mean :",x_mean)
y_mean=np.mean(y)
print("y.mean :",x_mean)
```

```python
## calculating the error
n=len(x)
       #y=b1x+b0 b1=slope,b0=intercept
numer=0
denom=0
for i in range(n):
    numer+=(x[i]-x_mean)*(y_mean)
    denom+=(x[i]-x_mean)**2
b1=numer/denom
b0=y_mean-(b1*x_mean)
print("slobe:",b1)
print("intercept :",b0)
```

```python
# y=mx+c
print("Linear Equation :y={:.5}x+{:.4}".format(b1,b0))
```

```python
y_predict=b1*x+b0
print(y_predict)
```

```python
## r2 score
n1=0
d1=0
for i in range(n):
    n1+=(y_predict[i]-y_mean)**2
    d1+=(y[i]-y_mean)**2
r2=1-n1/d1
print("r2 score:",r2)
```

```python
plt.scatter(x,y_predict,label="Linear wave",color="m")
plt.xlabel("x")
plt.ylabel("y_predicted value")
plt.title("Linear regresion with model includening")
plt.legend()
```
